Online retail analytics has become an important research area due to the rapid growth of e-commerce and the increasing availability of transactional customer data. Previous studies have shown that purchasing behavior, customer loyalty, product demand, and sales performance can be effectively analyzed using transaction-level retail datasets.

Research by Chen, Han, and colleagues demonstrated that association rule mining techniques such as Apriori and FP-Growth are effective for discovering relationships among products frequently purchased together. These techniques support market basket analysis, helping retailers improve product placement, cross-selling, and promotional strategies.

Fader, Hardie, and Lee (2005) introduced customer lifetime value models, emphasizing that customers differ significantly in purchasing frequency and monetary contribution. Their work highlighted the importance of identifying high-value customers for targeted marketing and long-term business profitability.

Research by Hughes (1994) proposed the Recency–Frequency–Monetary (RFM) framework for customer segmentation. Later studies confirmed that RFM analysis is an effective method for identifying loyal customers, inactive customers, and high-value customer groups, allowing businesses to personalize marketing campaigns and improve customer retention.

Studies by Agrawal and Srikant (1994) demonstrated that association rule mining provides valuable insights into purchasing patterns by identifying products frequently bought together. These findings have become the foundation of recommendation systems used by modern e-commerce platforms.

Previous research has also shown that online retail transaction data often contains missing customer identifiers, cancelled invoices, returned products, negative quantities, zero-priced items, and highly skewed purchasing behavior. Researchers emphasize that careful data cleaning and preprocessing are essential before applying clustering or association rule mining techniques.

Overall, the literature suggests that customer purchasing behavior is influenced by purchasing frequency, spending patterns, seasonal demand, product popularity, and geographical location. Exploratory Data Analysis (EDA) serves as the foundation for understanding these characteristics before implementing advanced machine learning and business intelligence techniques.

## Data Cleaning

The dataset was systematically cleaned to ensure high data quality and analytical reliability. The **InvoiceDate** column was converted to a datetime format, followed by a comprehensive assessment of missing values, duplicate records, and data consistency. Records with **missing product descriptions**, **duplicate transactions**, and **zero unit prices** were removed, while **cancelled invoices** and return-related transactions were excluded to retain only valid sales records. Product descriptions were standardized by removing extra whitespace and converting text to uppercase, and a **Revenue** feature was created by multiplying **Quantity** and **Price**. Finally, additional temporal features, including **Year, Month, Day, Weekday, and Hour**, were extracted to support advanced time-series and sales trend analysis.

````python
# -----------------------------
# 1. Use original Quantity (no filtering)
# -----------------------------
q = df['Quantity']

# -----------------------------
# 2. Create plot
# -----------------------------
plt.figure(figsize=(12,6))

sns.histplot(q, bins=50, kde=True, color='skyblue')

# -----------------------------
# 3. Mean & Median
# -----------------------------
mean_val = q.mean()
median_val = q.median()

plt.axvline(mean_val, color='red', linestyle='--', label=f"Mean = {mean_val:.2f}")
plt.axvline(median_val, color='green', linestyle='--', label=f"Median = {median_val:.2f}")

# -----------------------------
# 4. Title + Labels
# -----------------------------
plt.title("📦 Order Size Distribution (Quantity)")
plt.xlabel("Quantity per Order")
plt.ylabel("Number of Transactions")

# -----------------------------
# 5. Business Interpretation (STATES)
# -----------------------------
plt.text(0.6, 0.9,
         "Small Orders → 1–5 items\nMedium Orders → 5–20 items\nBulk Orders → 20+ items",
         transform=plt.gca().transAxes,
         fontsize=10,
         bbox=dict(facecolor='white', alpha=0.7))

plt.legend()
plt.show()
````

## 📌 Finding

Most customer orders contain only a small number of items (median = 4), while a few bulk purchases create significant outliers, resulting in a highly right-skewed order quantity distribution.

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# 1. Convert datetime
# -----------------------------
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# -----------------------------
# 2. Create Month
# -----------------------------
df['Month'] = df['InvoiceDate'].dt.month

# -----------------------------
# 3. Monthly Revenue
# -----------------------------
monthly_revenue = df.groupby('Month')['Revenue'].sum().reset_index()

# -----------------------------
# 4. Create figure (2 graphs side by side)
# -----------------------------
plt.figure(figsize=(15,6))

# -----------------------------
# GRAPH 1: Boxplot (Spread + Outliers)
# -----------------------------
plt.subplot(1,2,1)
sns.boxplot(x='Month', y='Revenue', data=df, palette='Set2')

plt.title("📦 Revenue Spread & Outliers by Month")
plt.xlabel("Month")
plt.ylabel("Revenue")

# -----------------------------
# GRAPH 2: Bar Chart (Performance)
# -----------------------------
plt.subplot(1,2,2)
sns.barplot(x='Month', y='Revenue', data=monthly_revenue, palette='viridis')

plt.title("📊 Monthly Revenue Performance")
plt.xlabel("Month")
plt.ylabel("Total Revenue")

# Add values on bars
for i, v in enumerate(monthly_revenue['Revenue']):
    plt.text(i, v, f"{int(v):,}", ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
````

## 📌 Findings

- Monthly revenue increased from **£1.07M in February** to a peak of **£2.97M in November**, before slightly decreasing to **£2.28M in December**.
- Most monthly revenues remain within a similar range, while a few transactions exceed **£150,000** and some returns reach around **-£55,000**, showing the presence of large orders and returns.

````python
# Convert date column
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"], errors="coerce")

# Create Revenue if not already present
if "Revenue" not in df.columns:
    df["Revenue"] = df["Quantity"] * df["Price"]

# Drop missing values
df = df.dropna(subset=["InvoiceDate", "Revenue"])

# Monthly revenue
monthly_revenue = df.set_index("InvoiceDate")["Revenue"].resample("M").sum()

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(x=monthly_revenue.index, y=monthly_revenue.values, marker="o", linewidth=2.5)

plt.title("Revenue by Month")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

## 📌 Findings

- Revenue increased from **approximately £0.82M in January 2010** to a peak of **around £1.50M in December 2011**, with noticeable growth during the last quarter of both years.
- After reaching its highest point in **December**, revenue dropped sharply to **about £0.64M in January 2012**, showing a clear year-end sales peak followed by a post-holiday decline.

````python
# -----------------------------
# 1. Convert datetime
# -----------------------------
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# -----------------------------
# 2. Extract Day & Hour
# -----------------------------
df['Day'] = df['InvoiceDate'].dt.day_name()
df['Hour'] = df['InvoiceDate'].dt.hour

# -----------------------------
# 3. Pivot Table (Revenue)
# -----------------------------
heatmap_data = df.pivot_table(
    values='Revenue',
    index='Day',
    columns='Hour',
    aggfunc='sum',
    fill_value=0
)

# -----------------------------
# 4. Order Days properly
# -----------------------------
order = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
heatmap_data = heatmap_data.reindex(order)

# -----------------------------
# 5. Plot Heatmap
# -----------------------------
plt.figure(figsize=(16,6))

sns.heatmap(
    heatmap_data,
    cmap='YlOrRd',
    linewidths=0.5,
    linecolor='white'
)

# -----------------------------
# 6. Title & Labels
# -----------------------------
plt.title("🔥 Revenue Heatmap (Day vs Hour) - Business Activity Pattern", fontsize=14)
plt.xlabel("Hour of Day")
plt.ylabel("Day of Week")

# -----------------------------
# 7. Add Insight Box (MOST IMPORTANT)
# -----------------------------
plt.text(0.02, -0.25,
         "💡 INSIGHTS:\n"
         "- Dark cells = High revenue (peak sales time)\n"
         "- Light cells = Low activity\n"
         "- Usually mid-day & evening hours show peak sales\n"
         "- Weekend vs Weekday behavior can be compared easily",
         transform=plt.gca().transAxes,
         fontsize=10,
         bbox=dict(facecolor='white', alpha=0.8))

plt.show()
````

## 📌 Findings

- The highest revenue is generated between **10:00 AM and 3:00 PM**, particularly on **Tuesday–Thursday**, whereas **Saturday** consistently records the lowest business activity.
- Revenue decreases after **5:00 PM**, showing that customer purchasing activity is concentrated during daytime business hours.

````python
# -----------------------------
# 1. Top customers by Revenue
# -----------------------------
top_customers = df.groupby('Customer ID')['Revenue'].sum().sort_values(ascending=False).head(10).reset_index()

# -----------------------------
# 2. Plot
# -----------------------------
plt.figure(figsize=(12,6))

sns.barplot(
    x='Customer ID',
    y='Revenue',
    data=top_customers,
    palette='viridis'
)

# -----------------------------
# 3. Labels & Title
# -----------------------------
plt.title("🏆 Top 10 Customers by Revenue (Customer Contribution)")
plt.xlabel("Customer ID")
plt.ylabel("Total Revenue")

# -----------------------------
# 4. Add values on bars (VERY PROFESSIONAL)
# -----------------------------
for i, v in enumerate(top_customers['Revenue']):
    plt.text(i, v, f"{int(v):,}", ha='center', va='bottom', fontsize=9)

plt.xticks(rotation=45)
plt.show()
````

- Revenue generally increases as the **order quantity increases**, with bulk orders (**20+ items**) contributing the highest sales, while most transactions contain fewer than **20 items**.

````python

# -----------------------------
# 1. Convert datetime
# -----------------------------
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# -----------------------------
# 2. Extract Year & Month
# -----------------------------
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month

# -----------------------------
# 3. Pivot Table (Revenue)
# -----------------------------
pivot_table = df.pivot_table(
    values='Revenue',
    index='Year',
    columns='Month',
    aggfunc='sum',
    fill_value=0
)

# -----------------------------
# 4. Heatmap
# -----------------------------
plt.figure(figsize=(12,6))

sns.heatmap(
    pivot_table,
    cmap='YlGnBu',
    linewidths=0.5,
    linecolor='white',
    annot=True,
    fmt=".0f"
)

plt.title("📊 Revenue Heatmap (Month vs Year) - Seasonality & Growth")
plt.xlabel("Month")
plt.ylabel("Year")

# -----------------------------
# 5. Business Insight Box
# -----------------------------
plt.text(
    0, -0.6,
    "💡 INSIGHTS:\n"
    "- Shows yearly growth trend\n"
    "- Identifies peak months (seasonality)\n"
    "- Helps compare performance across years\n"
    "- Dark cells = high revenue periods",
    transform=plt.gca().transAxes,
    fontsize=10,
    bbox=dict(facecolor='white', alpha=0.8)
)

plt.show()
````

## 📌 Findings

- Revenue shows a consistent year-end increase, with **November** recording the highest sales in both **2010 (£1.46M)** and **2011 (£1.50M)**, followed by strong performance in **October**.
- Most months generated **£0.5M–£0.9M** in revenue, while the period from **September to November** consistently delivered the strongest sales across both years, indicating a clear seasonal pattern.

````python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# Revenue by Country
country_rev = (
    df.groupby("Country")["Revenue"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

# Percentage contribution
percent = (country_rev / country_rev.sum() * 100)

plt.figure(figsize=(12,6))

bars = plt.barh(country_rev.index,
                country_rev.values,
                edgecolor='black')

plt.gca().invert_yaxis()

# Value labels
for i, (value, pct) in enumerate(zip(country_rev.values, percent)):
    plt.text(value,
             i,
             f"£{value:,.0f}\n({pct:.1f}%)",
             va='center',
             fontsize=10)

plt.title("Top 10 Countries by Total Revenue", fontsize=16, fontweight='bold')
plt.xlabel("Revenue (£)")
plt.ylabel("Country")

plt.gca().xaxis.set_major_formatter(
    mtick.FuncFormatter(lambda x, pos: f'£{x/1e6:.1f}M')
)

plt.grid(axis='x', linestyle='--', alpha=0.4)

plt.figtext(
    0.5, -0.05,
    "The United Kingdom dominates total revenue, while only a few international markets contribute significantly,\nindicating strong market concentration and opportunities for regional expansion.",
    ha='center',
    fontsize=10
)

plt.tight_layout()
plt.show()
````
